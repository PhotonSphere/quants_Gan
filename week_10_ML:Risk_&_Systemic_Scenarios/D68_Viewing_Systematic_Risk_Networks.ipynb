## ðŸ§© Bayesian Network for Systemic Risk Propagation

In this section, we extend the credit risk analysis into the realm of **systemic financial interactions** using a **Bayesian Network**.  
The goal is to model how shocks in one market (e.g., equity downturns) can propagate through related financial channels such as **foreign exchange (FX)**, **bond yields**, and **credit spreads**.

### ðŸŽ¯ Objectives
1. **Define Key Financial Nodes:**
   - **Equity Index:** Represents market valuation of firms and investor sentiment.
   - **FX Rate:** Captures currency value fluctuations that may result from market shocks.
   - **Bond Yields:** Reflect interest rate movements and monetary policy responses.
   - **Credit Spreads:** Represent risk premiums on debt instruments â€” key for credit risk assessment.

2. **Construct a Bayesian Network:**
   - Use directed edges to represent causal dependencies between financial variables.
   - For example:
     - Equity â†“ â†’ FX depreciation
     - Equity â†“ â†’ Bond yield shifts
     - Bond yields â†‘ â†’ Credit spreads widen

3. **Propagate a Shock Scenario:**
   - Introduce a simulated **â€“20% Equity Shock**.
   - Compute posterior probabilities of adverse moves in FX, bond yields, and credit spreads.
   - Compare these post-shock probabilities with baseline priors to assess systemic risk transmission.

4. **Visualize the Propagation Tree:**
   - Create a network diagram showing the probabilistic flow of stress across markets.
   - Highlight the most impacted nodes based on posterior probabilities.

### ðŸ§  Why Bayesian Networks?
Bayesian Networks provide a **transparent and probabilistic framework** for systemic risk modeling:
- They encode **causal relationships** explicitly.
- They allow **scenario-based inference** (e.g., stress testing).
- They combine **interpretability** (like Logistic Regression) with **dynamic propagation** (like complex simulation models).

This makes them ideal for credit and financial stability analysis, where **uncertainty**, **dependencies**, and **explainability** are crucial.

### ðŸ§  Using Real Financial Data

To extend this model for **empirical systemic risk analysis**, the Bayesian structure should be derived from actual market data rather than manually specified probabilities.

**Steps for real-data implementation:**

1. **Collect Historical Financial Data**
   - Equity indices (e.g., S&P 500, DAX, Nikkei)
   - FX rates (e.g., EUR/USD, USD/JPY)
   - Government bond yields (e.g., 10-Year Treasuries, Bund yields)
   - Credit spreads (e.g., corporate bond or CDS spreads)

2. **Data Preparation**
   - Convert continuous time series into **discrete categorical states**, such as:
     - â€œUpâ€, â€œStableâ€, â€œDownâ€ for equity and FX movements
     - â€œTightâ€, â€œStableâ€, â€œWidenâ€ for credit spreads
   - Align the frequency and timestamps of all series for consistent observations.

3. **Learn Network Structure**
   - Apply structure-learning algorithms (e.g., Hill Climb Search or PC Algorithm) from `pgmpy` to infer the causal dependencies among variables.

4. **Estimate Parameters**
   - Fit **Conditional Probability Distributions (CPDs)** from historical data using Maximum Likelihood Estimation (MLE) or Bayesian Estimation.

5. **Run Stress Scenarios**
   - Introduce shocks (e.g., equity drop or rate increase) as evidence.
   - Compute posterior probabilities to quantify how shocks propagate through financial channels.

```python
# ----------------------------
# Bayesian Network: Systemic Shock Propagation (pgmpy)
# ----------------------------

# !pip install pgmpy networkx matplotlib

from pgmpy.models import DiscreteBayesianNetwork as BayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import itertools


# ----------------------------
# 1) Define node states
# ----------------------------
# Equity (as % shock mapped to discrete state)
equity_states = ['SevereDown', 'ModerateDown', 'Stable', 'Up']  # SevereDown ~ -20%

# FX: direction of (domestic) currency vs benchmark: Depreciate is adverse for FX-sensitive assets
fx_states = ['Depreciate', 'Stable', 'Appreciate']

# Bond Yields: Fall (safe-haven), Stable, Rise (adverse if corporate yields rise; we treat 'Rise' as adverse for yields)
yield_states = ['Fall', 'Stable', 'Rise']

# Credit Spreads: Tight (good), Stable, Widen (adverse)
spread_states = ['Tight', 'Stable', 'Widen']

# ----------------------------
# 2) Build network structure (edges reflect causal assumptions)
# Equity -> FX, Equity -> BondYields, Equity -> CreditSpreads
# FX -> BondYields (FX moves can affect yields)
# BondYields -> CreditSpreads (yields affect spreads)
# ----------------------------
model = BayesianNetwork([
    ('Equity', 'FX'),
    ('Equity', 'BondYields'),
    ('Equity', 'CreditSpreads'),
    ('FX', 'BondYields'),
    ('BondYields', 'CreditSpreads')
])

# ----------------------------
# 3) Define CPDs (Tabular) -- illustrative expert-driven probabilities
# ----------------------------
# Prior on Equity (prior belief; can be changed)
equity_cpd = TabularCPD(
    variable='Equity',
    variable_card=len(equity_states),
    values=[[0.02],   # SevereDown
            [0.08],   # ModerateDown
            [0.75],   # Stable
            [0.15]],  # Up
    state_names={'Equity': equity_states}
)

# FX CPD conditioned on Equity
# Rows = FX states (Depreciate, Stable, Appreciate)
# Columns = Equity states in order [SevereDown, ModerateDown, Stable, Up]
fx_cpd = TabularCPD(
    variable='FX',
    variable_card=len(fx_states),
    values=[
        # Depreciate probabilities
        [0.85, 0.60, 0.20, 0.10],
        # Stable probabilities
        [0.12, 0.30, 0.70, 0.50],
        # Appreciate probabilities
        [0.03, 0.10, 0.10, 0.40]
    ],
    evidence=['Equity'],
    evidence_card=[len(equity_states)],
    state_names={'FX': fx_states, 'Equity': equity_states}
)

# Bond Yields CPD conditioned on Equity and FX
# Rows = BondYields (Fall, Stable, Rise)
# Columns correspond to cartesian product of Equity x FX states in the order pgmpy expects:
# For clarity we will manually build the table: for each Equity state, for each FX state -> probabilities
# We'll build a matrix with columns = len(Equity) * len(FX)
import itertools
cols = []
for e in equity_states:
    for f in fx_states:
        # intuition: equity severe down + FX depreciate -> risk-off -> yields FALL (safe-haven)
        if e == 'SevereDown' and f == 'Depreciate':
            cols.append([0.75, 0.20, 0.05])  # Fall most likely
        elif e == 'SevereDown' and f == 'Stable':
            cols.append([0.65, 0.30, 0.05])
        elif e == 'SevereDown' and f == 'Appreciate':
            cols.append([0.55, 0.35, 0.10])
        elif e == 'ModerateDown' and f == 'Depreciate':
            cols.append([0.60, 0.30, 0.10])
        elif e == 'ModerateDown' and f == 'Stable':
            cols.append([0.45, 0.45, 0.10])
        elif e == 'ModerateDown' and f == 'Appreciate':
            cols.append([0.35, 0.50, 0.15])
        elif e == 'Stable' and f == 'Depreciate':
            cols.append([0.30, 0.55, 0.15])
        elif e == 'Stable' and f == 'Stable':
            cols.append([0.10, 0.80, 0.10])
        elif e == 'Stable' and f == 'Appreciate':
            cols.append([0.05, 0.70, 0.25])
        elif e == 'Up' and f == 'Depreciate':
            cols.append([0.20, 0.60, 0.20])
        elif e == 'Up' and f == 'Stable':
            cols.append([0.02, 0.75, 0.23])
        elif e == 'Up' and f == 'Appreciate':
            cols.append([0.01, 0.60, 0.39])
        else:
            cols.append([0.1, 0.8, 0.1])
# transpose to create rows (each list above is column vector for (Fall,Stable,Rise))
values = list(zip(*cols))  # now a list of 3 tuples -> convert to lists
bond_cpd = TabularCPD(
    variable='BondYields',
    variable_card=len(yield_states),
    values=[list(map(float, values[i])) for i in range(len(values))],
    evidence=['Equity', 'FX'],
    evidence_card=[len(equity_states), len(fx_states)],
    state_names={'BondYields': yield_states, 'Equity': equity_states, 'FX': fx_states}
)

# Credit Spreads CPD conditioned on Equity and BondYields
# Rows = CreditSpreads (Tight, Stable, Widen)
cols_spread = []
for e in equity_states:
    for by in yield_states:
        # intuition: equity down & yields rise -> spreads widen (adverse)
        if e == 'SevereDown' and by == 'Rise':
            cols_spread.append([0.05, 0.20, 0.75])  # Widen most likely
        elif e == 'SevereDown' and by == 'Stable':
            cols_spread.append([0.10, 0.35, 0.55])
        elif e == 'SevereDown' and by == 'Fall':
            cols_spread.append([0.15, 0.45, 0.40])
        elif e == 'ModerateDown' and by == 'Rise':
            cols_spread.append([0.08, 0.30, 0.62])
        elif e == 'ModerateDown' and by == 'Stable':
            cols_spread.append([0.15, 0.45, 0.40])
        elif e == 'ModerateDown' and by == 'Fall':
            cols_spread.append([0.20, 0.55, 0.25])
        elif e == 'Stable' and by == 'Rise':
            cols_spread.append([0.25, 0.60, 0.15])
        elif e == 'Stable' and by == 'Stable':
            cols_spread.append([0.40, 0.55, 0.05])
        elif e == 'Stable' and by == 'Fall':
            cols_spread.append([0.60, 0.38, 0.02])
        elif e == 'Up' and by == 'Rise':
            cols_spread.append([0.50, 0.45, 0.05])
        elif e == 'Up' and by == 'Stable':
            cols_spread.append([0.65, 0.33, 0.02])
        elif e == 'Up' and by == 'Fall':
            cols_spread.append([0.78, 0.22, 0.00])
        else:
            cols_spread.append([0.3,0.6,0.1])

vals_spread = list(zip(*cols_spread))
spread_cpd = TabularCPD(
    variable='CreditSpreads',
    variable_card=len(spread_states),
    values=[list(map(float, vals_spread[i])) for i in range(len(vals_spread))],
    evidence=['Equity', 'BondYields'],
    evidence_card=[len(equity_states), len(yield_states)],
    state_names={'CreditSpreads': spread_states, 'Equity': equity_states, 'BondYields': yield_states}
)

# ----------------------------
# 4) Add CPDs to model and check
# ----------------------------
model.add_cpds(equity_cpd, fx_cpd, bond_cpd, spread_cpd)
assert model.check_model(), "CPD validation failed!"

# ----------------------------
# 5) Inference: prior marginals (without shock)
# ----------------------------
infer = VariableElimination(model)

def get_marginal(inference, var, states):
    marg = inference.query([var], show_progress=False)
    # updated for pgmpy â‰¥ 0.1.24
    s = marg.values
    return dict(zip(states, s))

prior_equity = get_marginal(infer, 'Equity', equity_states)
prior_fx = get_marginal(infer, 'FX', fx_states)
prior_yields = get_marginal(infer, 'BondYields', yield_states)
prior_spreads = get_marginal(infer, 'CreditSpreads', spread_states)

print("===== PRIOR MARGINALS =====")
print("Equity prior:", prior_equity)
print("FX prior:", prior_fx)
print("Bond Yields prior:", prior_yields)
print("Credit Spreads prior:", prior_spreads)

# ----------------------------
# 6) Propagate shock: Equity = SevereDown (representing -20%)
# ----------------------------
evidence = {'Equity': 'SevereDown'}
posterior_fx = infer.query(['FX'], evidence=evidence, show_progress=False).values
posterior_yields = infer.query(['BondYields'], evidence=evidence, show_progress=False).values
posterior_spreads = infer.query(['CreditSpreads'], evidence=evidence, show_progress=False).values


posterior_fx = dict(zip(fx_states, posterior_fx))
posterior_yields = dict(zip(yield_states, posterior_yields))
posterior_spreads = dict(zip(spread_states, posterior_spreads))

print("\n===== POSTERIOR MARGINALS GIVEN Equity = SevereDown (â‰ˆ -20%) =====")
print("FX posterior:", posterior_fx)
print("Bond Yields posterior:", posterior_yields)
print("Credit Spreads posterior:", posterior_spreads)

# ----------------------------
# 7) Create a propagation visualization (tree) showing prior vs posterior of ADVERSE states
# ----------------------------
# For visual simplicity, define 'adverse' state per node:
adverse_state = {
    'Equity': 'SevereDown',
    'FX': 'Depreciate',
    'BondYields': 'Rise',       # treat Rise as adverse for corporate funding
    'CreditSpreads': 'Widen'
}

# Collect prior and posterior adverse probabilities
nodes = ['Equity', 'FX', 'BondYields', 'CreditSpreads']
prior_probs = {
    'Equity': prior_equity[adverse_state['Equity']],
    'FX': prior_fx[adverse_state['FX']],
    'BondYields': prior_yields[adverse_state['BondYields']],
    'CreditSpreads': prior_spreads[adverse_state['CreditSpreads']]
}
posterior_probs = {
    'Equity': 1.0 if evidence['Equity'] == adverse_state['Equity'] else prior_probs['Equity'],
    'FX': posterior_fx[adverse_state['FX']],
    'BondYields': posterior_yields[adverse_state['BondYields']],
    'CreditSpreads': posterior_spreads[adverse_state['CreditSpreads']]
}

# Build directed graph for plotting
G = nx.DiGraph()
G.add_edges_from([('Equity','FX'), ('Equity','BondYields'), ('Equity','CreditSpreads'),
                  ('FX','BondYields'), ('BondYields','CreditSpreads')])

pos = {
    'Equity': (0, 0.6),
    'FX': (-1.0, 0.0),
    'BondYields': (0.0, 0.0),
    'CreditSpreads': (1.0, 0.0)
}

plt.figure(figsize=(10,5))
ax = plt.gca()
# node colors by post-adverse probability intensity (red intensity)
node_colors = []
for n in nodes:
    p = posterior_probs[n]
    # map probability to color intensity (low->light, high->red)
    node_colors.append((1.0, 1.0 - p, 1.0 - p))  # redish where p close to 1

nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=20, edge_color='gray', width=2, ax=ax)
nx.draw_networkx_nodes(G, pos, node_size=2500, node_color=node_colors, ax=ax)
# labels: include name + prior->posterior adverse probability
labels = {}
for n in nodes:
    labels[n] = f"{n}\nprior adverse: {prior_probs[n]:.2%}\npost adverse: {posterior_probs[n]:.2%}"
nx.draw_networkx_labels(G, pos, labels=labels, font_size=10, font_weight='bold', ax=ax)

plt.title("Systemic Shock Propagation Tree\n(Equity -20% â†’ posterior adverse probabilities)")
plt.axis('off')
plt.show()

# ----------------------------
# 8) Tabular summary of change in adverse probability
# ----------------------------
summary = pd.DataFrame({
    'Node': nodes,
    'AdverseState': [adverse_state[n] for n in nodes],
    'PriorAdverseProb': [prior_probs[n] for n in nodes],
    'PosteriorAdverseProb': [posterior_probs[n] for n in nodes]
})
summary['Delta'] = summary['PosteriorAdverseProb'] - summary['PriorAdverseProb']
print("\n===== SUMMARY: Adverse State Probabilities (prior vs posterior) =====")
print(summary)
```

- **Bond yields**
- **Credit spreads**

A simulated **â€“20% equity shock** (classified as â€œSevereDownâ€) was introduced, and its impact on the other nodes was evaluated through posterior probability updates.  
By comparing **prior** and **posterior** marginal distributions, we observe how stress transmits across interconnected financial channels.

These priors reflect a calm financial environment with low systemic stress.

---

### ðŸ”º Posterior Probabilities (Given Equity = SevereDown)
After applying an **equity market shock (â‰ˆ â€“20%)**, the Bayesian network propagates updated probabilities through its dependencies:
